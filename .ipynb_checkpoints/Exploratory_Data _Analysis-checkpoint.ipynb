# Barplots of Asset Managers with based on the amount of investments

```python

def make_barplots(data_list,titles_list, names_list):
    
    count_investments = {}
    index = 0
    row = 0
    col = 0
    f, ax = plt.subplots(ncols= 2, nrows = 2, figsize = (20, 15))
    f.suptitle("Total number of investments held by Asset Managers", fontsize = 30)
    sns.set_theme(style="darkgrid")
    sns.despine(left=True, bottom=True)
    plt.subplots_adjust(wspace = 0.8)
    
    for data_dict in data_list:
        
        for cik in data_dict.keys():
            issuers_dict = data_dict.get(cik)
            count_investments[names_list[index].get(cik)] = len(issuers_dict)
        
        df = pd.DataFrame(list(count_investments.items()), columns = ['name', 'investments'])
        df = df.sort_values('investments', ascending = False)
        

        # Initialize the matplotlib figure
        
        g = sns.barplot(x= 'investments', y= 'name', palette="viridis", data = df.iloc[0:20,], ax = ax[row, col])
        if (col ==1): row = 1
        col = 1 if col == 0 else 0
        g.title.set_text(titles_list[index])
        index += 1
    

        
    
```

```python
data_list = [data2021, data2020, data2019, data2018]
names_list = [names2021, names2020, names2019, names2018]

titles_list = ['2021', '2020', '2019', '2018']
make_barplots(data_list, titles_list, names_list)
```

# Lineplot of 

```python
def make_lineplot(data_list, names_list):
    f, ax = plt.subplots(figsize = (10, 10))
    f.suptitle("Total number of investments held by Asset Managers")
    sns.set_theme(style="darkgrid")
    sns.despine(left=True, bottom=True)
    #years_list = [2021, 2020, 2019, 2018]
    years_list = [1, 2, 3, 4, 5]
    index = 0
    final = pd.DataFrame(columns = ['name', 'counts', 'year'])
    name_occurences = {}
    
    for data_dict in data_list:
        count_investments = {}

        for cik in data_dict.keys():
            if names_list[index].get(cik) in name_occurences.keys():
                name_occurences[names_list[index].get(cik)] += 1 
            else:
                name_occurences[names_list[index].get(cik)] = 1
            issuers_dict = data_dict.get(cik)
            count_investments[names_list[index].get(cik)] = len(issuers_dict)
    
        df = pd.DataFrame(list(count_investments.items()), columns = ['name', 'counts'])
        df["year"] = np.repeat(years_list[index],df.shape[0], axis = 0)
        df.counts = df.counts.astype(float)
        df.year = df.year.astype(int)
        final = pd.concat([df, final])
        index += 1
        
    complete = pd.DataFrame(columns = ['name', 'counts', 'year'])
    for i in range(0, final.shape[0]):
        if name_occurences.get(final.iloc[i, 0]) == 5:
            row = list(final.iloc[i, ])
            print(row)
            complete = complete.append(pd.Series(row, index=['name', 'counts', 'year']), ignore_index=True)



    
    sns.lineplot(data=complete, x = "year", y="counts", hue = "name")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
   
            
        
        
```

```python
def make_lineplot(data_list):
    f, ax = plt.subplots(figsize = (10, 10))
    f.suptitle("Total number of investments held by Asset Managers")
    sns.set_theme(style="darkgrid")
    sns.despine(left=True, bottom=True)
    df = pd.DataFrame(columns = ["cik", "name", "quarter", "amount"])
    common_ciks = set(data_list[0].keys())
    
    for data_dict in data_list[1:]:
        cik = set(data_dict.keys())
        common_ciks = common_ciks.intersection(cik)
        
    print(common_ciks)
    for cik in common_ciks:
        my_dict = {}
        my_dict["quarter"] = [2018, 2019, 2020, 2021]
        my_dict["cik"] = np.repeat(cik, 4 , axis =0)
        my_dict["name"] = np.repeat(names_dict.get(cik),4 , axis = 0)
        arr = []
        arr.append(len(data_list[3].get(cik)) if data_list[3].get(cik) != None else 0)
        arr.append(len(data_list[2].get(cik)) if data_list[2].get(cik) != None else 0)
        arr.append(len(data_list[1].get(cik)) if data_list[1].get(cik) != None else 0)
        arr.append(len(data_list[0].get(cik)) if data_list[0].get(cik) != None else 0)
        my_dict["amount"] = arr
        df2 = pd.DataFrame.from_dict(my_dict)
        df2.quarter = df2.quarter.astype(int)
        df2.amount = df2.amount.astype(float)
        df = pd.concat([df, df2])

    sns.lineplot(data=df, x = "quarter", y="amount", hue = "name")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
   
    
```
