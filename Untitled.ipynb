```python
# making Dictionaries to be used for visualization

    most_investmennts = {} # count of investments per investmet manager
    most_stocks_cik = {} # sum of total stocks per issuer 
    most_stocks_issuer = {} # amount of stocks invested in company
    most_investors = {} # 

    added_issuers = set()

    for cik in data_dict.keys():
        issuers_dict = data_dict.get(cik)
        most_investments[names_dict.get(cik)] = len(issuers_dict)
        most_stocks_cik[names_dict.get(cik)] = 0
        for issuer in issuers_dict.keys():
            most_stocks_cik[names_dict.get(cik)] += issuers_dict.get(issuer)
            if (issuer in added_issuers):
                most_stocks_issuer[issuer] += issuers_dict.get(issuer)
                most_investors[issuer] +=  1
            else: 
                most_stocks_issuer[issuer] = issuers_dict.get(issuer)
                most_investors[issuer] =  1
                added_issuers.add(issuer)
```
